```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cluster_mpg.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df_dummies=pd.get_dummies(df.drop('name',axis=1),dtype=int)
```

```python
df_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaled_data=scaler.fit_transform(df_dummies)
```

```python
scaled_data      # range -> (0,1)
```

```python
scaled_df=pd.DataFrame(scaled_data,columns=df_dummies.columns)
```

```python
scaled_df
```

```python
#  visualizing relationships
plt.figure(figsize=(10,7),dpi=200)
sns.heatmap(data=scaled_df)
```

```python
plt.figure(figsize=(10,7),dpi=200)
sns.clustermap(data=scaled_df,row_cluster=False)
```
