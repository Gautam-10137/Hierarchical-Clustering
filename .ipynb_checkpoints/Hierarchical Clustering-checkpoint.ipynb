```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cluster_mpg.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df_dummies=pd.get_dummies(df.drop('name',axis=1),dtype=int)
```

```python
df_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaled_data=scaler.fit_transform(df_dummies)
```

```python
scaled_data      # range -> (0,1)
```

```python
scaled_df=pd.DataFrame(scaled_data,columns=df_dummies.columns)
```

```python
scaled_df
```

```python
#  visualizing relationships
plt.figure(figsize=(10,7),dpi=200)
sns.heatmap(data=scaled_df)
```

```python
plt.figure(figsize=(10,7),dpi=200)
sns.clustermap(data=scaled_df,row_cluster=False)
```

# Now using sci-kit learn

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels=model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
# visualizing the distribution of clusters in original data
sns.scatterplot(data=df,x='mpg',y='horsepower',hue=cluster_labels,palette='viridis')
```

```python
#  max distance possible? --> sqrt(N_features)
```

```python
len(scaled_df.columns)
```

```python
np.sqrt(10)   # Max distance possible between two points (Min/Max Scaling)
```

```python
model=AgglomerativeClustering(n_clusters=None,distance_threshold=0) 
# no. of clusters=0 and max distance for linkage -> 0 so no two points can merge , so it will take every point as new cluster
```

```python
cluster_labels= model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
#  to get the mathematical information used to form dendrogram  -> we use linkage matrix
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix= hierarchy.linkage(model.children_)
```

```python
linkage_matrix    #[point1,point2,distance b/w them,no. of pointsin this cluster]
```

```python
# plotting dendogram
```

```python
plt.figure(figsize=(12,9))
dendro=dendrogram(linkage_matrix)
```

```python
#  truncating these cluster representation for better visualization
```

```python
#  Truncation is used to condense the dendrogram
plt.figure(figsize=(12,9))
dendro=dendrogram(linkage_matrix,truncate_mode='lastp',p=10)
```

```python
plt.figure(figsize=(12,9))
dendro=dendrogram(linkage_matrix,truncate_mode='level',p=3)
```
